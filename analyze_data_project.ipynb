## Does Germany Has a higher Life Expectancy at Birth than United States?

```python
print(df.head())
```

```python
df.Country.unique()
```

```python
## Let's first start by exploring how long their datasets are, and how should we visualize them.

germany = df.loc[df['Country'] == 'Germany']
united_states = df.loc[df['Country'] == 'United States of America']
```

```python
len(united_states), len(germany)
```

```python
germany['Life expectancy at birth (years)'].mean(), germany['Life expectancy at birth (years)'].std()
```

```python
united_states['Life expectancy at birth (years)'].mean(), united_states['Life expectancy at birth (years)'].std()
```

```python
united_states['Life expectancy at birth (years)'][round(0.1 * len(united_states)) : round(0.9 * len(united_states))].mean(), united_states['Life expectancy at birth (years)'][round(0.1 * len(united_states)) : round(0.9 * len(united_states))].std()
```

```python
germany['Life expectancy at birth (years)'][round(0.1 * len(germany)) : round(0.9 * len(germany))].mean(), germany['Life expectancy at birth (years)'][round(0.1 * len(germany)) : round(0.9 * len(germany))].std()
```

```python
# They look similar! But, what about if we do a hypotesis test to see if they are different?
# Hypothesis Null: Germany and United States of America have the same life expectancy
# Hypothesis Alternative: Germany and United States of America have different life expectancy

from scipy.stats import ttest_ind
ttest_ind(germany['Life expectancy at birth (years)'], united_states['Life expectancy at birth (years)']) # We assumed that their variances are the same!
```

```python
# Now, let'not assume that their variances are the same, let's assume that they are not the same

ttest_ind(germany['Life expectancy at birth (years)'], united_states['Life expectancy at birth (years)'], equal_var=False)
```

```python
# It seems that there wasn't any change, and the p-value stays almost the same!
# Now, what should we infer about the p-value? 
# The p-value is 2.52e-05, which is less than 0.05.
# Sinnce we choose 0.05 as our common threshold, we can reject the null hypothesis.
# Therefore, we can conclude that Germany and United States of America seems to have different life expectancy.
```

```python
plt.hist(united_states['Life expectancy at birth (years)'], bins=30)
```

```python
# We can also use the Mann-Whitney U test to see if they are different!

from scipy.stats import mannwhitneyu

mannwhitneyu(germany['Life expectancy at birth (years)'], united_states['Life expectancy at birth (years)'])
```

```python
# It looks that we can reject the null hypothesis again!
# Therefore, we can conclude that Germany and United States of America seems to have different life expectancy.
```

```python
# Let's create several models for every country since they have a great correlation with Life Expectancy and the log of GDP!

import statsmodels.api as sm

df = pd.DataFrame({'X': np.log(df.GDP), 'y': df['Life expectancy at birth (years)']})

X_with_constant = sm.add_constant(df['X'])

model = sm.OLS(df['y'], X_with_constant).fit()

results = model.summary()

print(results)
```

```python
predicted_y = model.predict(X_with_constant)

plt.scatter(df.X, df.y)
plt.plot(df.X, predicted_y, color='r')
```

```python
# Let's try  to predict now for every country!

model_per_country = {}

results_per_country = {}

for country in countries:
    df = pd.DataFrame({'X': np.log(data_per_country.get_group(country).GDP), 'y': data_per_country.get_group(country)['Life expectancy at birth (years)']})

    X_with_constant = sm.add_constant(df['X'])

    model = sm.OLS(df['y'], X_with_constant).fit()

    model_per_country[country] = model

    results = model.summary()

    results_per_country[country] = results
    
    print(results)
```

```python
# First, let's see the scale of GDP to understand how should we increase it!

df.describe()
```

```python
# What about if we increase by 1.0e12 a data point and see what happens?

df.groupby('Country').GDP.max()
```

```python
# Let's take the max point and increase it by 1.0e12

new_points = df.groupby('Country').GDP.max() + 1.0e12 # Basically we are adding a billion to each GDP for each country
```

```python
new_point_usa = new_points['United States of America']
new_point_gm = new_points['Germany']

new_point_usa, new_point_gm
```

```python
def create_new_data_points(new_point, quantity, increase_by = None, multiply_by = None):

    new_points = []

    if isinstance(increase_by, float) and multiply_by is None:
        for _ in range(quantity):

            new_points.append(new_point + increase_by * np.random.random())

            new_point += increase_by * np.random.random()

    elif isinstance(multiply_by, float) and increase_by is None:

        for _ in range(quantity):

            new_points.append(new_point * multiply_by * np.random.random())

            new_point *= multiply_by * np.random.random()

    else:
        raise ValueError('Please specify either increase_by or multiply_by')

    return np.array(new_points)

new_points_usa = create_new_data_points(new_point_usa, 20, increase_by=1.0e12)
new_points_gm = create_new_data_points(new_point_gm, 20, increase_by=1.0e12)

new_points_usa, new_points_gm
```

```python
# We have created new data with some noise in it, now let's add them to our original dataset

usa_points = df[df.Country == 'United States of America'].GDP

print(len(usa_points))

usa_points = np.append(usa_points, new_points_usa)

gm_points = df[df.Country == 'Germany'].GDP

print(len(usa_points))

gm_points = np.append(gm_points, new_points_gm)

len(usa_points), len(gm_points)
```

```python
# Now, let's try to predict the values in a separated way and then we will plot them together

new_points_predicted_usa = model_per_country['United States of America'].predict(sm.add_constant(np.log(new_points_usa)))
new_points_predicted_gm = model_per_country['Germany'].predict(sm.add_constant(np.log(new_points_gm)))

new_points_predicted_usa, new_points_predicted_gm
```

```python
# Let's try to plot the together now!


sns.scatterplot(x=np.log(usa_points), y=np.append(y_predict_per_country['United States of America'], new_points_predicted_usa))
sns.scatterplot(x=np.log(gm_points), y=np.append(y_predict_per_country['Germany'], new_points_predicted_gm))
plt.legend(['United States of America', 'Germany'])
```
